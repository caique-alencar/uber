```python
import json
import pandas

with open('credenciais_uber.json') as cred_data:
    info = json.load(cred_data)
    server_token = info['server_token']
    
# Define credencial de acesso
from uber_rides.session import Session
session = Session(server_token)
from uber_rides.client import UberRidesClient
client = UberRidesClient(session)

# Bloco define a latitude e a longitude do local de partida e do destino
response = client.get_price_estimates(
    start_latitude=-22.894925, 
    start_longitude=-43.294089,
    end_latitude=-22.941335,  
    end_longitude=-43.202743,
    seat_count=2
)
estimate = response.json.get('prices')

# Exibe os dicionários com os dados brutos das estimativas
# for item in estimate:
    # print(item)
    # print()

planilha = []

# Bloco coleta só as informações relevantes e exibe de forma mais amigável
cont = 0
# Enquanto o contador for menor que quantidade de itens(dicionários) da lista estimate
while cont < len(estimate):
    # Para cada dicionário da lista estimate, salvar valor de chave desejada
    for item in estimate:
        tipo_uber = estimate[cont]["display_name"]
        distancia = estimate[cont]["distance"]
        duracao = estimate[cont]["duration"]
        estimativa = estimate[cont]["estimate"]
    # Exibe os valores coletados para cada tipo de viagem no Uber
    print(f"Opção {cont+1} de corrida")
    print(f"Tipo de Uber: {tipo_uber}")
    print(f"Distância da corrida: {distancia}km")
    print(f"Duração da corrida: {duracao}")
    print(f"Estimativa de preço: {estimativa}")
    print()
    cont+=1
    
    dicionario = {"Tipo de Uber": tipo_uber, "Distância da corrida": distancia, 
                  "Duração da corrida": duracao, "Estimativa de preço": estimativa}
    planilha.append(dicionario)
    
# Exibição dos dados de forma mais bonita usando df do pandas
df = pandas.DataFrame(planilha)
df.sort_values(by = "Estimativa de preço", ascending = False)
```
